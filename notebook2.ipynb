## ðŸ§ Why Use Bash in Software Engineering?

When developing a web application that involves creating files and directories, as well as searching for text in text files using shell scripts, a suitable scripting language to use is **Bash (Bourne Again Shell)**. Bash is widely used in Unix-based systems like Linux and macOS, offering file manipulation, directory operations and text processing. Bash scripts are portable across Unix-based systems, making them a versatile choice for automating tasks and system administration across different environments.

Bash provides a rich set of commands for:

- Creating files and directories
- Moving, copying and deleting files
- Writing text and searching text within files

Commands such as:

| Task | Bash Command |
|------|--------------|
| Create folders | `mkdir` |
| Create files | `touch` |
| Write/append text | `echo`, `>`, `>>` |
| Display file contents | `cat` |
| Search text patterns | `grep` |
| List directories as a tree | `tree` |

Later in this notebook, you will also create **executable shell scripts (`.sh`)** that automate these actions.

### ðŸ’» Creating a Bash Script

Create a file ending in `.sh`, then add commands to it.

#### Example:
```bash
mkdir logs
touch logs/app.log
echo "Application started" >> logs/app.log
grep "Application" logs/app.log
```

Run it:
```bash
bash setup.sh
```

````bash
%%bash
echo "#!/bin/bash" > setup.sh
echo "mkdir -p logs" >> setup.sh
echo "touch logs/app.log" >> setup.sh
echo "echo 'Application started' >> logs/app.log" >> setup.sh
echo "grep 'Application' logs/app.log" >> setup.sh

bash setup.sh
tree logs
````
