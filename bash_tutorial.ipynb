## Introduction: Why Bash for Web Development?

When developing a web application that involves creating files and directories, as well as searching for text in text files using shell scripts, a suitable scripting language to use is Bash (Bourne Again Shell). Bash is a widely used shell scripting language in Unix-based systems like Linux and macOS, offering file manipulation, directory operations and text processing. Bash scripts are portable across Unix-based systems, making them a versatile choice for automating tasks and system administration across different environments.

Bash provides a rich set of commands and utilities for creating files and directories, moving, copying, and deleting files, as well as setting permissions and ownership. Commands like `mkdir` for creating directories, `touch` for creating files, `mv` for moving files, `cp` for copying files, and `rm` for deleting files are commonly used in Bash scripts. Bash offers powerful text processing capabilities, allowing you to search for specific text patterns in files, extract data, and manipulate text content. Commands like `grep` for searching text in files based on patterns, `sed` for text substitution and manipulation, and `awk` for text processing and reporting are commonly used for text processing tasks.

Bash enables seamless integration with system commands and utilities, enabling developers to combine shell commands with file operations and text processing within the script. Developers can execute system commands, capture their output and process the data using Bash scripting.

### Getting Started

- **Mac/Linux users**: Open Terminal application

### 10. grep - Search Text in Files

Searches for text patterns in files.

**Syntax:**
```bash
grep "pattern" filename
grep -i "pattern" filename  # Case-insensitive
grep -n "pattern" filename  # Show line numbers
grep -r "pattern" directory/  # Recursive search
grep -v "pattern" filename  # Invert match (show non-matching lines)
grep -c "pattern" filename  # Count matches
```

````bash
%%bash
# Create a sample log file
cat > logfile.txt << EOF
2024-03-15 INFO Application started
2024-03-15 ERROR Database connection failed
2024-03-15 INFO User logged in
2024-03-15 ERROR Invalid credentials
2024-03-15 SUCCESS Operation completed
EOF

# Basic search
echo "Lines containing ERROR:"
grep "ERROR" logfile.txt

# Search with line numbers
echo "\nWith line numbers:"
grep -n "ERROR" logfile.txt

# Count matches
echo "\nCount of errors:"
grep -c "ERROR" logfile.txt
````

## Part 2: Introduction to Shell Scripting (15 minutes)

### What is a Shell Script?

A shell script is a text file containing a series of commands that the shell can execute. Scripts automate repetitive tasks and combine multiple commands into a single executable program.

### Creating Your First Script

**Key Script Components:**

1. **Shebang Line:** `#!/bin/bash` - Tells the system to use Bash
2. **Comments:** `#` - Lines starting with # are comments
3. **Commands:** Any valid Bash commands
4. **Variables:** Store and reuse values

````bash
%%bash
# Create and run a simple script
cat > my_first_script.sh << 'EOF'
#!/bin/bash
# This is a comment - my first script

echo "Hello from my shell script!"
echo "Today's date is:"
date
echo "Current directory:"
pwd
EOF

# Make it executable
chmod +x my_first_script.sh

# Run the script
./my_first_script.sh
````

### Using Variables in Scripts

````bash
%%bash
cat > variables_demo.sh << 'EOF'
#!/bin/bash
# Demonstrating variables

NAME="Software Engineering"
YEAR=2024
SCHOOL="HSC"

echo "Subject: $NAME"
echo "Year: $YEAR"
echo "Program: $SCHOOL"
echo "Full: $SCHOOL $NAME $YEAR"
EOF

chmod +x variables_demo.sh
./variables_demo.sh
````

### Practical Script Example: Website Setup

````bash
%%bash
cat > website_setup.sh << 'EOF'
#!/bin/bash
# Script to set up a basic website structure

PROJECT_NAME="my_website"

# Create directory structure
echo "Creating project structure..."
mkdir -p $PROJECT_NAME/html
mkdir -p $PROJECT_NAME/css
mkdir -p $PROJECT_NAME/js
mkdir -p $PROJECT_NAME/images

# Create initial files
echo "Creating files..."
touch $PROJECT_NAME/html/index.html
touch $PROJECT_NAME/html/about.html
touch $PROJECT_NAME/css/styles.css
touch $PROJECT_NAME/js/script.js

# Add basic HTML template
cat > $PROJECT_NAME/html/index.html << HTMLEOF
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>My Website</title>
    <link rel="stylesheet" href="../css/styles.css">
</head>
<body>
    <h1>Welcome to My Website</h1>
    <script src="../js/script.js"></script>
</body>
</html>
HTMLEOF

echo "Website structure created successfully!"

# Display structure
if command -v tree &> /dev/null; then
    tree $PROJECT_NAME
else
    ls -R $PROJECT_NAME
fi
EOF

chmod +x website_setup.sh
./website_setup.sh
````

### Script with Text Search

````bash
%%bash
cat > find_todos.sh << 'EOF'
#!/bin/bash
# Script to find TODO comments in code files

# Create sample files with TODOs
mkdir -p src
echo "// TODO: Implement login function" > src/app.js
echo "<!-- TODO: Add navigation -->" > src/index.html
echo "function test() { /* TODO: Add tests */ }" > src/utils.js

SEARCH_DIR="src"
SEARCH_TERM="TODO"

echo "Searching for $SEARCH_TERM comments..."
echo "================================"

# Search for TODO in all JavaScript and HTML files
grep -rn "$SEARCH_TERM" $SEARCH_DIR --include="*.js" --include="*.html"

# Count total TODOs
TODO_COUNT=(grep -r "
SEARCH_TERM" $SEARCH_DIR --include="*.js" --include="*.html" | wc -l)
echo "================================"
echo "Total TODOs found: $TODO_COUNT"
EOF

chmod +x find_todos.sh
./find_todos.sh
````
