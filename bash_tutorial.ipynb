# Bash basics: `mkdir`, `touch`, `tree`, `cat`

This notebook demonstrates how to run bash commands inside notebook cells. Use **Run** on each cell to execute the commands. Cells that start with `%%bash` run in a bash shell. You can also use `!command` in a Python cell for single-line commands.

## Notes for students
- Files created in the Binder session are ephemeral (they disappear when the session stops). Download anything you want to keep.
- Use `%%bash` at the top of a cell to run multiple shell commands in that cell.


```python
%%bash
echo "Make a directory called exercise1 and create two files inside it"
rm -rf exercise1
mkdir -p exercise1
touch exercise1/file1.txt exercise1/file2.txt
echo "This is file1" > exercise1/file1.txt
echo "This is file2" > exercise1/file2.txt
echo "Contents of exercise1:" 
ls -l exercise1
```

```python
%%bash
echo "Show tree view of current directory"
tree -a -I '.ipynb_checkpoints' || true
```

### Exercise: create a folder `practice` with subfolders and files

1. Create `practice/docs` and `practice/src`
2. Create files `practice/docs/readme.md` and `practice/src/main.sh` with some text
3. Display the tree for `practice` and show the content of `practice/docs/readme.md`

Write the commands in the cell below and run it.

```python
%%bash
# student should edit/run these commands to complete the exercise
rm -rf practice
mkdir -p practice/docs practice/src
echo "# Practice readme\nThis is a practice readme." > practice/docs/readme.md
echo "#!/bin/bash\necho Hello from main" > practice/src/main.sh
chmod +x practice/src/main.sh
echo "Tree view of practice:"
tree practice
echo "\nContents of practice/docs/readme.md:"
cat practice/docs/readme.md
```

### Quick tips
- Use `%%bash` for multi-line shell scripts inside a cell.
- Use `!ls` or `!cat filename` from a Python cell for a single command.
- If `tree` fails to run, ensure `apt.txt` is present in the repository with `tree` listed so Binder installs it at build time.
